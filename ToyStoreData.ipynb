```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
inventory = pd.read_csv('/content/inventory.csv')
product = pd.read_csv('/content/products.csv')
sales = pd.read_csv('/content/sales.csv')
stores = pd.read_csv('/content/stores.csv')
```

```python
inventory
```

```python
inventory.info()
```

```python
product['Product_Cost'] = product['Product_Cost'].str.replace('$', '').astype(float)
product['Product_Price'] = product['Product_Price'].str.replace('$', '').astype(float)
product
```

```python
product.info()
```

```python
sales['Date'] = pd.to_datetime(sales['Date'])
sales
```

```python
sales.info()
```

```python
stores['Store_Open_Date'] = pd.to_datetime(stores['Store_Open_Date'])
stores
```

```python
stores.info()
```

```python
sales_prod = sales.merge(product, on='Product_ID')
sales_prod
```

```python
sales_prod['Total_Cost'] = sales_prod['Units'] * sales_prod['Product_Cost']
sales_prod['Total_Price'] = sales_prod['Units'] * sales_prod['Product_Price']
sales_prod['Total_Profit'] = (sales_prod['Total_Price'] - sales_prod['Total_Cost']).round(2)
sales_prod['Total_Profit'] = sales_prod['Total_Profit'].apply(lambda x: '{:.2f}'.format(x))
sales_prod
```

```python
sales_prod['Total_Profit'] = sales_prod['Total_Profit'].astype(float)
sales_prod.groupby('Product_Name')['Total_Profit'].sum().sort_values(ascending=False)
```

```python
sales_prod.groupby('Product_Category')['Total_Profit'].sum().sort_values(ascending=False)
```

```python
sales_prod.groupby('Product_Category')['Total_Profit'].sum().sort_values(ascending=False).plot(kind='bar', title = 'Total Profit by Category')
```

```python
product_inv = product.merge(inventory, on='Product_ID')
product_inv
```

```python
product_inv.groupby('Product_Name')['Stock_On_Hand'].sum().sort_values(ascending=False)
product_inv['Cost_On_Hand'] = product_inv['Product_Cost'] * product_inv['Stock_On_Hand']
product_inv['Price_On_Hand'] = product_inv['Product_Price'] * product_inv['Stock_On_Hand']
product_inv['Potential_Profit'] = product_inv['Price_On_Hand'] - product_inv['Cost_On_Hand']
product_inv['Potential_Profit'] = product_inv['Potential_Profit'].apply(lambda x: '{:.2f}'.format(x))
product_inv
```

```python
product_inv['Potential_Profit'] = product_inv['Potential_Profit'].astype(float)
product_inv.groupby('Product_Name')['Potential_Profit'].sum().sort_values(ascending=False).plot(kind = 'bar')
```

```python
product_inv.groupby('Product_Name')['Cost_On_Hand'].sum().sort_values(ascending=False).plot(kind = 'bar')
```

```python
product_inv.groupby('Product_Name')['Price_On_Hand'].sum().sort_values(ascending=False).plot(kind = 'bar')
```

```python
sales_store = sales.merge(stores, on='Store_ID')

sales_store
```

```python
sales_store_prod = sales_store.merge(product, on='Product_ID')
sales_store_prod
```

```python
sales_store_prod['Cost_Per_Store'] = sales_store_prod['Product_Cost'] * sales_store_prod['Units']
sales_store_prod['Price_Per_Store'] = sales_store_prod['Product_Price'] * sales_store_prod['Units']
sales_store_prod['Profit_Per_Store'] = sales_store_prod['Price_Per_Store'] - sales_store_prod['Cost_Per_Store']
sales_store_prod
```

```python
sales_store_prod.groupby('Store_Name')['Profit_Per_Store'].sum().sort_values(ascending=False)
```

```python
sales_store_prod.groupby('Store_Name')['Profit_Per_Store'].sum().sort_values(ascending= False).plot(kind = 'bar',
                                                                      figsize = (12, 6))
```

```python
maven_cdmx2_sales = sales_store_prod[sales_store_prod['Store_Name'] == 'Maven Toys Ciudad de Mexico 2']
```

```python
maven_cdmx2_sales = sales_store_prod[sales_store_prod['Store_Name'] == 'Maven Toys Ciudad de Mexico 2']
maven_cdmx2_sales
```

```python
maven_cdmx2_sales_by_date = maven_cdmx2_sales.groupby('Date')['Units'].sum().reset_index()
maven_cdmx2_sales_by_date
```

```python
plt.figure(figsize=(12, 6))
plt.plot(maven_cdmx2_sales_by_date['Date'], maven_cdmx2_sales_by_date['Units'])
plt.title('Sales of maven Toys Cidudad de Mexico 2 by Date')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.grid(True)
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
filtered_sales = maven_cdmx2_sales_by_date[
    (maven_cdmx2_sales_by_date['Date'] >= '2018-01-01') & (maven_cdmx2_sales_by_date['Date'] <= '2018-07-31')
]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(filtered_sales['Date'], filtered_sales['Units'])
plt.title('Sales of Maven Toys Ciudad de Mexico 2 by Date (2018-01 to 2018-07)')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
